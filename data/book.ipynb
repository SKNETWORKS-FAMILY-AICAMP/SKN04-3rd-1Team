```python
if not books:
    print("도서 정보를 확인할 수 없습니다.")
else:
    processed_books = []
    korean_regex = re.compile("^[가-힣]")

    for book in books:
        title = book.get("title")
        # audienceNote = book.get("audienceNote")
        
        if not title or not korean_regex.match(title):
            continue
        
        # if audienceNote != "일반이용자":
        #     continue
        
        kdc_code = book.get("kdc")
        main_kdc_code = kdc_code[:2] if kdc_code else None 
        main_kdc_description = kdc_categories.get(main_kdc_code)
        
        ddc_code = book.get("ddc")
        main_ddc_code = ddc_code[:2] if ddc_code else None
        main_ddc_description = ddc_categories.get(main_ddc_code)
        
        if main_kdc_description is None and main_ddc_description is None:
            continue

        
        processed_book = {
            "title": book.get("title", "없음"),
            "creator": book.get("creator", "없음"),
            "publisher": book.get("publisher", "없음"),
            "issued": book.get("issued", "없음"),
            "kdc": book.get("kdc", "없음"),
            "ddc": book.get("ddc", "없음"),
            "mainCategoryDescription": main_kdc_description if main_kdc_description else main_ddc_description,
            "alternative": book.get("alternative", "없음"),
            "keyword": book.get("keyword", "없음"),
            "seeAlso": book.get("seeAlso", "없음"),
        }

        processed_books.append(processed_book)
```
